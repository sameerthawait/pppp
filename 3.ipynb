```python
class Job:
    def __init__(self, id, deadline, profit):
        self.id = id
        self.deadline = deadline
        self.profit = profit

def merge(arr, l, m, r):
    L = arr[l:m+1]
    R = arr[m+1:r+1]

    i = j = 0
    k = l

    while i < len(L) and j < len(R):
        if L[i].profit > R[j].profit:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    while i < len(L):
        arr[k] = L[i]
        i += 1
        k += 1

    while j < len(R):
        arr[k] = R[j]
        j += 1
        k += 1

def mergeSort(arr, l, r):
    if l < r:
        m = (l + r) // 2
        mergeSort(arr, l, m)
        mergeSort(arr, m+1, r)
        merge(arr, l, m, r)

def jobSequencing(arr, n):
    mergeSort(arr, 0, n-1)

    slot = [0]*100
    result = [0]*100
    totalProfit = 0

    for i in range(n):
        for j in range(arr[i].deadline - 1, -1, -1):
            if slot[j] == 0:
                slot[j] = 1
                result[j] = arr[i].id
                totalProfit += arr[i].profit
                break

    print("Selected Jobs:", end=" ")
    for i in range(n):
        if slot[i]:
            print(result[i], end=" ")

    print("\nTotal Profit:", totalProfit)

# Input
n = int(input("Enter number of jobs: "))
arr = []

print("Enter Job ID, Deadline and Profit:")
for _ in range(n):
    id, deadline, profit = map(int, input().split())
    arr.append(Job(id, deadline, profit))

jobSequencing(arr, n)
```
