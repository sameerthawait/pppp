```python
class Article:
    def __init__(self, profit, weight):
        self.profit = profit
        self.weight = weight

def fractionalKnapsack(W, arr):
    arr.sort(key=lambda x: x.profit / x.weight, reverse=True)

    curWeight = 0
    finalprofit = 0.0

    for item in arr:
        if curWeight + item.weight <= W:
            curWeight += item.weight
            finalprofit += item.profit
        else:
            remain = W - curWeight
            finalprofit += item.profit / item.weight * remain
            break

    return finalprofit

# Input
n = int(input("Enter number of items: "))
W = int(input("Enter capacity of knapsack: "))

arr = []
print("Enter profit and weight for each item:")
for _ in range(n):
    p, w = map(int, input().split())
    arr.append(Article(p, w))

ans = fractionalKnapsack(W, arr)

print("The maximum profit is", round(ans, 2))
```
