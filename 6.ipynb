```python
from itertools import permutations

minCost = float('inf')
bestPath = []

def tsp(graph):
    global minCost, bestPath
    n = len(graph)

    cities = list(range(1, n))

    for perm in permutations(cities):
        currentCost = 0
        k = 0
        path = [0]

        for city in perm:
            currentCost += graph[k][city]
            k = city
            path.append(city)

        currentCost += graph[k][0]
        path.append(0)

        if currentCost < minCost:
            minCost = currentCost
            bestPath = path

# Example graph (same as your Java code)
graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

tsp(graph)

print("Optimal Path:", bestPath)
print("Minimum Cost:", minCost)
```
