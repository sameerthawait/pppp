```python
import heapq

class Node:
    def __init__(self, level, profit, weight, bound, items):
        self.level = level
        self.profit = profit
        self.weight = weight
        self.bound = bound
        self.items = items[:]

    def __lt__(self, other):
        return self.bound > other.bound   # max heap behavior

# Calculate upper bound
def bound(u, n, W, weights, profits):
    if u.weight >= W:
        return 0

    profit_bound = u.profit
    j = u.level + 1
    total_weight = u.weight

    while j < n and total_weight + weights[j] <= W:
        total_weight += weights[j]
        profit_bound += profits[j]
        j += 1

    if j < n:
        profit_bound += (W - total_weight) * (profits[j] / weights[j])

    return profit_bound

def knapsack(W, weights, profits, n):

    # Sort by profit/weight ratio
    index = list(range(n))
    index.sort(key=lambda i: profits[i] / weights[i], reverse=True)

    sortedW = [weights[i] for i in index]
    sortedP = [profits[i] for i in index]

    pq = []
    u = Node(-1, 0, 0, 0, [])
    u.bound = bound(u, n, W, sortedW, sortedP)
    heapq.heappush(pq, u)

    maxProfit = 0
    bestItems = []

    while pq:
        u = heapq.heappop(pq)

        if u.bound > maxProfit and u.level < n - 1:

            # Include item
            v = Node(u.level + 1,
                     u.profit + sortedP[u.level + 1],
                     u.weight + sortedW[u.level + 1],
                     0,
                     u.items + [index[u.level + 1]])

            if v.weight <= W and v.profit > maxProfit:
                maxProfit = v.profit
                bestItems = v.items

            v.bound = bound(v, n, W, sortedW, sortedP)
            if v.bound > maxProfit:
                heapq.heappush(pq, v)

            # Exclude item
            v = Node(u.level + 1, u.profit, u.weight, 0, u.items)
            v.bound = bound(v, n, W, sortedW, sortedP)

            if v.bound > maxProfit:
                heapq.heappush(pq, v)

    print("Maximum Profit:", maxProfit)
    print("Items selected:", bestItems)


# Example 1
profits1 = [60, 100, 120]
weights1 = [10, 20, 30]
W1 = 50

knapsack(W1, weights1, profits1, len(profits1))

# Example 2
profits2 = [40, 35, 60, 55]
weights2 = [4, 3, 5, 6]
W2 = 10

knapsack(W2, weights2, profits2, len(profits2))
```
